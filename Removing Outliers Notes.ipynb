```python
# Standard Deviation Method for Normal Distribution with outliers that are far from mean.
#summary statistics
#still includes outliers
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def remove_outliers_std(data,threshold=2):
    mean = np.mean(data)
    std_dev = np.std(data)
    lower_bound = mean - threshold * std_dev
    upper_bound = mean + threshold * std_dev
    return [x for x in data if lower_bound <= x <= upper_bound]
data = pd.read_csv('C:/Users/hninshwezinhlaing/Downloads/CRM+Sales+Opportunities/sales_pipeline.csv')
sales = data[data['close_value'] > 0]
filter_data = remove_outliers_std(sales['close_value'])
print(filter_data)
plt.boxplot(filter_data)
plt.show()
```

```python
#remove_outliers_percentile for skewed data or data with extreme outliers
def remove_outliers_percentile(data, lower_percentile = 25, upper_percentile = 75):
    lower_bound = np.percentile(data, lower_percentile)
    upper_bound = np.percentile(data,upper_percentile)
    return [x for x in data if lower_bound <= x <= upper_bound]
filtered_data_percentile = remove_outliers_percentile(filter_data)
print(filtered_data_percentile)
plt.boxplot(filtered_data_percentile)
plt.show()
```

```python
#boxplot method for identifying outliers visually and removing them based on the whisker of the boxplot
def remove_outliers_boxplot(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data,75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return [x for x in data if lower_bound <= x <= upper_bound]
filtered_data_boxplot = remove_outliers_boxplot(filter_data)
print(filtered_data_boxplot)
plt.boxplot(filtered_data_boxplot)
plt.show()
```
